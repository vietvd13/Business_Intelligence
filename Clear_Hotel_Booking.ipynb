```python
import numpy as np
import csv

hotel_bookings = []

for row in csv.reader(open('hotel_bookings.csv', 'r'), delimiter=','):
    hotel_bookings.append(row)
```

```python
header = hotel_bookings[0] # Get column headers
hotel_bookings = hotel_bookings[1:] # Get data
```

```python
# Clear col 'Cancel'
def fixCancel(row):
    isCancel = int(row[1])
    if isCancel == 0:
        row[1] = 'No Cancel'
    else:
        row[1] = 'Cancel'
    return row

hotel_bookings = list(map(fixCancel, hotel_bookings))
```

```python
# Merge arrival (Format: Year-Month-Date)
def mergeArrival(row):
    arrival = [row[3], row[4], row[6]]
    key = '-'
    row[3] = key.join(arrival)
    return row
    
hotel_bookings = list(map(mergeArrival, hotel_bookings))
```

```python
# Merge 'children, babies'
def mergeChildren(row):
    count = row[10] + row[11]
    row[10] = count
    return row

hotel_bookings = list(map(mergeChildren, hotel_bookings))
```
